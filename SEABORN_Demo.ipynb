```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
#COORELATION WITH HEATMAP
```

```python
df.dtypes
```

```python
corr = df[['total_bill' , 'tip' , 'size']].corr()
```

```python
sns.heatmap(corr)
```

```python
#JOINPLOT
```

```python
sns.jointplot(x='tip', y = 'total_bill' , data = df , kind = 'hex')
```

```python
sns.jointplot(x='total_bill', y = 'tip' , data = df , kind = 'reg')
```

```python
#PAIRPLOT
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='day')
```

```python
#DISTPLOT
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False , bins=10 , color = 'red')
```
