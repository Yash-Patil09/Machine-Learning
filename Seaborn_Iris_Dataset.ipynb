```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
import matplotlib.pyplot as plt
```

```python
df.dtypes
```

```python
df[['sepal_length','sepal_width','petal_length','petal_width']].corr()
```

```python
corr = df[['sepal_length','sepal_width','petal_length','petal_width']].corr()
```

```python
sns.heatmap(corr)
```

```python
#JOINTPLOT
```

```python
sns.jointplot(x='sepal_length' , y='sepal_width', data = df, hue='species' , kind='scatter')
```

```python
sns.jointplot(x='petal_length' , y='petal_width', data = df, hue='species' , kind='scatter')
```

```python
#PAIRPLOT
```

```python
sns.pairplot(df, hue='species')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
#HISTPLOT
```

```python
sns.histplot(df['sepal_length'] , kde=False ,bins=10 , color='red', edgecolor='black')
```

```python
sns.countplot(x='species',data=df,palette='muted',edgecolor='black')
```

```python
#BARPLOT
```

```python
sns.barplot(x='species',y='sepal_length',data=df,palette='muted',edgecolor='black')
```

```python
sns.barplot(x='species',y='sepal_width',data=df,palette='muted',edgecolor='black')
```

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Bar plot for Sepal Width
sns.barplot(ax=axes[0], x="species", y="sepal_width", data=df, palette="muted", edgecolor="black")
axes[0].set_title("Average Sepal Width by Species")
axes[0].set_xlabel("Species")
axes[0].set_ylabel("Sepal Width")


# Bar plot for Sepal Length
sns.barplot(ax=axes[1], x="species", y="sepal_length", data=df, palette="muted", edgecolor="black")
axes[1].set_title("Average Sepal Length by Species")
axes[1].set_xlabel("Species")
axes[1].set_ylabel("Sepal Length")

plt.tight_layout()
plt.show()
```
